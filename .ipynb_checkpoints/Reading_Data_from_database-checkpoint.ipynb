```python
# SQLAlchemy
import sqlalchemy
import psycopg2
import datetime as dt
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine , inspect , func
from flask import Flask

db_password = 'postgres'
```

```python
# Path to sqlite
database_path = f"postgresql://postgres:{db_password}@127.0.0.1:5432/Consumer_complaints_db"

# Create an engine that can talk to the database
engine = create_engine(database_path)
```

```python
# Query All Records in the the Database
#data = engine.execute("SELECT * FROM complaints")

inspector = inspect(engine)
inspector.get_table_names()
```

```python
columns = inspector.get_columns('complaints')
for c in columns:
    print(c['name'], c["type"])
```

```python
engine.execute('SELECT * FROM complaints').fetchall()
```

```python
# Reflect Database into ORM class
Base = automap_base()
Base.prepare(autoload_with=engine)
Complaints = Base.classes.complaints
```

```python
session = Session(engine)
```

```python
sel1 = [Complaints.product, func.count(Complaints.product)]
data1 = session.query(*sel1).group_by(Complaints.product).all()
data1
```

```python
sel2 = [Complaints.submitted_via, func.count(Complaints.submitted_via)]
data2 = session.query(*sel2).group_by(Complaints.submitted_via).all()
data2
```

```python
app = Flask(__name__)
```

```python
@app.route("/")
def index() :
    return " Project 3"

@app.route("/Productdata")
def Productdata() :
    return data1

@app.route("/Submitteddata")
def Submitteddata() :
    return data2
```

```python
if __name__ == "__main__":
    app.run(debug=True)
```
